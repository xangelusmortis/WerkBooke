```python
("""Name: Joshua Shadle
    Date: 3/173
    Purpose: This code is made to show you how many seconds there are in a complete day.
    The bottom plays arouns with the different types and division methods that Pyhton has to offer.""")

minutes_per_hour = 60
seconds_per_minute = 60
hours_per_day = 24

print("how many seconds are there per hour?")
seconds_per_hour = minutes_per_hour * seconds_per_minute
print(seconds_per_hour, "seconds per hour.\n")

print("How many seconds are in a day?")
seconds_per_day = hours_per_day * seconds_per_hour
print(seconds_per_day, "seconds per day.\n")

print("Floating point division:")
print(seconds_per_day/seconds_per_hour,"\n")

print("Integer division:")
print(seconds_per_day//seconds_per_hour)
```
